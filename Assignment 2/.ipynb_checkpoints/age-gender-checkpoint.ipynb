```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Literacy Total.csv')

df.head(50)
```

```python
df1 = pd.read_csv('Bi and Tri Literacy2.csv', usecols=['State Code' , 'State Name' , 'Educational level' , 'Three Male' ,'Three Female'])


df1.head()
```

```python
df2 = df.merge(df1,on=['State Code' , 'Educational level'],how='inner')

df2['male_ratio'] = np.round((df2['Three Male'] / df2['TOT_M']),4)
df2['female_ratio'] = np.round((df2['Three Female'] / df2['TOT_F']),4)

df2.head(50)
```

```python
data ={}
df4 = pd.DataFrame()

for i in range(36):
    
    df3 = df2[df2['State Code'] == i]
    
    data['State Code'] = i
    v1 = df3.iloc[df3.male_ratio.argmax(),0:].to_numpy().tolist()
    v2 = df3.iloc[df3.female_ratio.argmax(),0:].to_numpy().tolist()
    
    data['male_ratio'] = v1[9]
    data['male_literacy-group'] = v1[2]
    data['female_ratio'] = v2[10]
    data['female_literacy-group'] = v2[2]
    
    df4 =df4.append(data,ignore_index = True)
    
df4.head(50)

df4.to_csv('literacy-gender-3.csv',index = False) 
```
